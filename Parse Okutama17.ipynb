```python
import os
import random
import numpy as np
#import xml.etree.ElementTree as ET

txt_path = '2.2.5.txt'

action_list = ["Walking","Sitting","Standing","Running","Lying","Carrying","Entering Cars/Buildings",\
               "Leaving Cars/Buildings", "Pushing/Pulling Something", "Reading", "Drinking", "Calling",\
               "Hand Shaking", "Hugging", "Delivering", "Pushing Someone",""]
action_dict = dict(zip(action_list, range(len(action_list))))

f = open(txt_path,'r')
annotations = []
actions = []
full_anns = []


for line in f:
    line = line.replace("\n","")
    anno = np.array([int(x) for x in line.split(' ')[0:8]])
    annotations.append(anno)
    full_anno = line.split(' ')
    full_anns.append(full_anno)
    line = line.replace(' "','"')
    action_set = []
    #print line
    for act in line.split('"')[3:]:
        if action_dict[act] != 16:
            action_set.append(action_dict[act])
    actions.append(action_set)

    #print anno
```

```python
   

annotations = np.array(annotations)
actions = np.array(actions)
frame_num = max(annotations[:,5])
print frame_num
annotation_pair = {}
for i in range(0,frame_num+1):
    #print annotations[annotations[:,5]==i]
    #print actions[annotations[:,5]==i]
    annotation_pair[i] = [annotations[annotations[:,5]==i],actions[annotations[:,5]==i]]



print "generating " + txt_path + " to xml format"
xml_path = txt_path.replace('.txt','')




for idx in xrange(1,frame_num+1):
    if random.randint(1,50) > 1:
        continue
    obj_num = len(annotation_pair[idx][0])
    print  "<annotation>"
    print  "<folder>", "images/"+txt_path.split('t')[0], "</folder>"
    print  "<filename>", str(idx) +".jpg", "</filename>"
    print  "<size>"
    print  "<width>", "1366", "</width>"
    print  "<height>", "768", "</height>"
    print  "<depth>", "3", "</depth>"
    print  "</size>"
    for i in xrange(0,obj_num):
        if annotation_pair[idx][0][i][6] == 1:
            continue
        print  "<object>"
        print  "<name>", str(annotation_pair[idx][1][i][:])[1:-1], "</name>"
        print  "<pose>Unspecified</pose>"
        print  "<truncated>",str(annotation_pair[idx][0][i][7]),"</truncated>"
        print  "<difficult>0</difficult>"              
        print  "<bndbox>"
        print  "<xmin>", str(annotation_pair[idx][0][i][1]), "</xmin>"
        print  "<xmax>", str(annotation_pair[idx][0][i][2]), "</xmax>"
        print  "<ymin>", str(annotation_pair[idx][0][i][3]), "</ymin>"
        print  "<ymax>", str(annotation_pair[idx][0][i][4]), "</ymax>"
        print  "</bndbox>"
        print  "</object>"   
    print  "</annotation>" 
```

```python
for idx in xrange(1,frame_num+1):
    if not os.path.exists(os.path.dirname(os.path.join(xml_path, str(idx) + '.xml'))):
            os.makedirs(os.path.dirname(os.path.join(xml_path,str(idx) + '.xml')))
    with open(os.path.join(xml_path, str(idx) + '.xml'), 'w') as f:
        obj_num = len(annotation_pair[idx][0])
        print >>f, "<annotation>"
        print >>f, "<folder>", "images/"+xml_path, "</folder>"
        print >>f, "<filename>", str(idx) +".jpg", "</filename>"
        print >>f, "<size>"
        print >>f, "<width>", "1366", "</width>"
        print >>f, "<height>", "768", "</height>"
        print >>f, "<depth>", "3", "</depth>"
        print >>f, "</size>"
        for i in xrange(0,obj_num):
            if annotation_pair[idx][0][i][6] == 1:
                continue
            print >>f, "<object>"
            print >>f, "<name>", str(annotation_pair[idx][1][i][:])[1:-1], "</name>"
            print >>f, "<pose>Unspecified</pose>"
            print >>f, "<truncated>",str(annotation_pair[idx][0][i][6]),"</truncated>"
            print >>f, "<difficult>0</difficult>"              
            print >>f, "<bndbox>"
            print >>f, "<xmin>", str(annotation_pair[idx][0][i][1]), "</xmin>"
            print >>f, "<xmax>", str(annotation_pair[idx][0][i][2]), "</xmax>"
            print >>f, "<ymin>", str(annotation_pair[idx][0][i][3]), "</ymin>"
            print >>f, "<ymax>", str(annotation_pair[idx][0][i][4]), "</ymax>"
            print >>f, "</bndbox>"
            print >>f, "</object>"
        print >>f, "</annotation>"      
```
